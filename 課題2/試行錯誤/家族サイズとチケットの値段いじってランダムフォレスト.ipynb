```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("train.csv")
df = df.drop(["PassengerId","Cabin"], axis = 1)
df = df.replace("male",0).replace("female",1).replace("S",0).replace("C",1).replace("Q",2)
df.count()
```

```python
df
y = df["Survived"].as_matrix()
```

```python
split_data = []
for survived in [0,1]:
    split_data.append(df[df.Survived==survived])

temp = [i["Age"].dropna() for i in split_data]
plt.hist(temp, histtype="barstacked", bins=20)
```

```python
temp = [i["SibSp"].dropna() for i in split_data]
plt.hist(temp, histtype="barstacked", bins=20,range=[2,8])
```

```python
FamilyisBig1 = []
for i in df["SibSp"]:
    if i > 2:
        FamilyisBig1.append(1)
    else:
        FamilyisBig1.append(0)

df['FamilyisBig1'] = np.array(FamilyisBig1)
```

```python
temp = [i["Parch"].dropna() for i in split_data]
plt.hist(temp, histtype="barstacked", bins=20,range=[3,6])
```

```python
FamilyisBig2 = []
for i in df["Parch"]:
    if i > 2:
        FamilyisBig2.append(1)
    else:
        FamilyisBig2.append(0)

df['FamilyisBig2'] = np.array(FamilyisBig2)
```

```python
temp = [i["Fare"].dropna() for i in split_data]
plt.hist(temp, histtype="barstacked", bins=20)
```

```python
ChicketisCheap = []
for i in df["Fare"]:
    if i < 50:
        ChicketisCheap.append(1)
    else:
        ChicketisCheap.append(0)

df['ChicketisCheap'] = np.array(ChicketisCheap)
```

```python
df
```

```python
df.count()
```

```python
df1 = df.drop(["Survived","Name","Ticket"], axis = 1)
df1 = df1.fillna(df1.mean())
X = df1.as_matrix()
df1.count()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
gbct = GradientBoostingClassifier(random_state=0, max_depth=3, learning_rate=0.1)
gbct.fit(X_train, y_train)
```

```python
print('train accuracy: %.3f' % gbct.score(X_train, y_train))
print('test  accuracy: %.3f' % gbct.score(X_test, y_test))
```

```python
df2 = pd.read_csv("test.csv")
df2 = df2.drop(["PassengerId","Cabin"], axis = 1)
df2 = df2.replace("male",0).replace("female",1).replace("S",0).replace("C",1).replace("Q",2)

    
FamilyisBig11 = []
for i in df2["SibSp"]:
    if i > 2:
        FamilyisBig11.append(1)
    else:
        FamilyisBig11.append(0)

df2['FamilyisBig1'] = np.array(FamilyisBig11)

FamilyisBig22 = []
for i in df2["Parch"]:
    if i > 2:
        FamilyisBig22.append(1)
    else:
        FamilyisBig22.append(0)

df2['FamilyisBig2'] = np.array(FamilyisBig22)

ChicketisCheap1 = []
for i in df2["Fare"]:
    if i < 50:
        ChicketisCheap1.append(1)
    else:
        ChicketisCheap1.append(0)

df2['ChicketisCheap'] = np.array(ChicketisCheap1)

df2 = df2.drop(["Name","Ticket","Fare","SibSp","Parch"], axis = 1)
df2 = df2.fillna(df2.mean())
X1 = df2.as_matrix()
```

```python
y1 = gbct.predict(X1)
```

```python
a =[i for i in range(892,1310)]
dfnew2 = pd.DataFrame({'PassengerId':a ,'Survived': y1})
dfnew2.to_csv("pred3.csv") 
```

```python
from sklearn.ensemble import RandomForestClassifier
forest = RandomForestClassifier(min_samples_leaf=3, random_state=0)
forest.fit(X_train, y_train)
print('Train score: {}'.format(forest.score(X_train, y_train)))
print('Test score: {}'.format(forest.score(X_test, y_test)))
```

```python
y2 = forest.predict(X1)
a =[i for i in range(892,1310)]
dfnew2 = pd.DataFrame({'PassengerId':a ,'Survived': y2})
dfnew2.to_csv("pred4.csv") 
```
